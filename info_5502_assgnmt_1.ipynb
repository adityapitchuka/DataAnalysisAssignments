```
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

#file_reader = open("/content/sample_data/Installation_Readme.txt","r")
file_reader = open("/content/sample_data/testing.txt","r")
list_strings = []

for line in file_reader:
  list_strings += line.split()

file_reader.close()

#list_strings = ['tom','jerry','and','$am','#am','*ing']

#Using Insertion Sort to sort the strings

for i in range(len(list_strings)):
	check_value = list_strings[i]
	check_pos = i
	while (check_pos > 0 and list_strings[check_pos-1] > check_value):
		list_strings[check_pos], list_strings[check_pos-1] = list_strings[check_pos-1], list_strings[check_pos]
		check_pos -= 1
	list_strings[check_pos] = check_value

#print("Sorted String List:")
print (list_strings)

#print(len(list_strings))

strings_lens = []
for i in range(len(list_strings)):
  strings_lens.append(len(list_strings[i]))

#print("Lengths each string in Sorted List:")
#print(strings_lens)


strings_order = [x+1 for x in range(len(list_strings))]

graph = Table().with_columns(['Sorted String Index',strings_order,'Length of Strings',strings_lens])
graph.scatter('Sorted String Index','Length of Strings')
graph.plot('Sorted String Index','Length of Strings')
```
