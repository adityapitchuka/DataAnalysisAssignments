```
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
import numpy as np
import random
import math


def pi_value_estimator(iteration,pi_values,interations_seq):

  count_inside = 0
  in_X, in_Y, out_X,out_Y = [], [], [], []
  
  interations_seq.append(iteration)

  #determining the x and y values by using random generator
  for count in range(0, iteration):
      #x = random.random()
      #y = random.random()
      x = 2*(random.random()-0.5)
      y = 2*(random.random()-0.5)
      d = math.sqrt(x**2 + y**2)  #using standard equation of a circle
      if d < 1:                   #if values are less than radius, we append to list of inside points
        count_inside += 1
        in_X.append(x)
        in_Y.append(y)
      else:                       #else we add to the outside list
        out_X.append(x)
        out_Y.append(y)
  count += 1
  pi_value = (4.0 * count_inside / count)

  pi_values.append(pi_value)

  #definiing a square
  rectangle = plt.Rectangle((-1,-1),2,2,ec="red",fill=False)
  #defining a circle
  circle = plt.Circle((0,0),1,ec="blue",fill=False)

  plt.gca().add_patch(rectangle)
  plt.gca().add_patch(circle)

  #plotting the points which are inside and outside
  plt.scatter(in_X,in_Y,color='#7EFB00',marker=".")
  plt.scatter(out_X,out_Y,color='#EFE43D',marker=".")

  plt.axis("equal")
  plt.show()

  return pi_value

pi_values = []
interations_seq = []

print("For 10 values, Estimated Pi Value is "+str(pi_value_estimator(10,pi_values,interations_seq)))
print("For 100 values, Estimated Pi Value is "+str(pi_value_estimator(100,pi_values,interations_seq)))
print("For 500 values, Estimated Pi Value is  "+str(pi_value_estimator(500,pi_values,interations_seq)))
print("For 1000 values, Estimated Pi Value is  "+str(pi_value_estimator(1000,pi_values,interations_seq)))
print("For 50000 values, Estimated Pi Value is  "+str(pi_value_estimator(50000,pi_values,interations_seq)))
print("For 100000 values, Estimated Pi Value is  "+str(pi_value_estimator(100000,pi_values,interations_seq)))
print("For 500000 values, Estimated Pi Value is  "+str(pi_value_estimator(500000,pi_values,interations_seq)))
```
