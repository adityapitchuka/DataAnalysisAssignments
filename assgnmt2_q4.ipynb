```
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

#Reading Table from data set
data_reader = Table.read_table('/content/sample_data/data_with_continents.csv')

#selecting 1990 Estimated HIV% 
scatter_1990 = data_reader.select('Continents','Country - HIV %','1990')
values_1990 = list(scatter_1990.column('1990'))

#selecting 2010 Estimated HIV%
scatter_2010 = data_reader.select('Continents','Country - HIV %','2010')
values_2010 = list(scatter_2010.column('2010'))

#list of continents
list_continents = scatter_1990.column('Continents')

#Plotting a scatter graph for 1990 Estimated HIV% values
plots.scatter(list_continents,values_1990,color='b')
plots.show()

print('\n')

#Plotting a scatter graph for 2010 Estimated HIV% values
plots.scatter(list_continents,values_2010,color='r')
plots.show()
```
