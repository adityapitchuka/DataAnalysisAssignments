```
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

#reading the dataset obtained from Q.1
data_reader = Table.read_table('/content/sample_data/data_with_continents.csv')   

#converting Table in to pandas data-frame
pandas_DF=data_reader.to_df() 

#filtering continent-wise, first for 'Africa'
africa_DF = pandas_DF.loc[pandas_DF['Continents'] == 'Africa']  

#finding the average for the columns
avg_africa  = round(africa_DF.iloc[:,4:35].mean(axis=0), 4) 

#plotting a graph
avg_africa.plot(legend='TRUE',label='Africa')  
plots.show()

print('\n')

#the same wille be repeated for all the rest of continents

#for Asia

asia_DF = pandas_DF.loc[pandas_DF['Continents'] == 'Asia']
avg_asia = round(asia_DF.iloc[:,4:35].mean(axis=0), 4)
avg_asia.plot(legend='TRUE',label='Asia')
plots.show()

print('\n')

#for Oceania

oceania_DF = pandas_DF.loc[pandas_DF['Continents'] == 'Oceania']
avg_oceania = round(oceania_DF.iloc[:,4:35].mean(axis=0), 4)
avg_oceania.plot(legend='TRUE',label='Oceania')
plots.show()
print('\n')

#for Europe

europe_DF = pandas_DF.loc[pandas_DF['Continents'] == 'Europe']
avg_europe = round(europe_DF.iloc[:,4:35].mean(axis=0), 4)
avg_europe.plot(legend='TRUE',label='Europe')
plots.show()
print('\n')


#for North America

NA_DF = pandas_DF.loc[pandas_DF['Continents'] == 'North America']
avg_NA  = round(NA_DF.iloc[:,4:35].mean(axis=0), 4)
avg_NA.plot(legend='TRUE',label='North America')
plots.show()
print('\n')

#for South America

SA_DF = pandas_DF.loc[pandas_DF['Continents'] == 'South America']
avg_SA = round(SA_DF.iloc[:,4:35].mean(axis=0), 4)
avg_SA.plot(legend='TRUE',label='South America')
plots.show()
print('\n')

#creating an overlaid graph using each of average found above 
avg_africa.plot(legend='TRUE',label='Africa')
avg_asia.plot(legend='TRUE',label='Asia')
avg_europe.plot(legend='TRUE',label='Europe')
avg_oceania.plot(legend='TRUE',label='Oceania')
avg_NA.plot(legend='TRUE',label='North America')
avg_SA.plot(legend='TRUE',label='South America')
plots.show()
```
